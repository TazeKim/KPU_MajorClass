```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('covid_19_data.csv',encoding='euc-kr')
data2 = pd.read_csv('time_series_covid_19_confirmed.csv',encoding='euc-kr')
```

```python
####################1번#########################

#날짜별로 전세계 발병 증가 변동 추이
date_data = data1.groupby(["ObservationDate"]).sum()
date_data = date_data[["Confirmed", "Deaths","Recovered"]]
date_data
```

```python
#날짜별 전 세계 코로나 발병 증가 변동 추이
plt.rc('font',family = 'Malgun Gothic')
all1_cht = date_data.plot(legend=True,figsize=(18, 8), title='날짜별 전 세계 코로나 발병 증가 변동 추이')
all1_cht.set_ylabel('Confirmed', fontsize=12)
```

```python
#날짜별 전 세계 코로나 사망자,완치자 증가 변동 추이
date_data_dth_rcb = date_data[["Deaths","Recovered"]]
plt.rc('font',family = 'Malgun Gothic')
all2_cht = date_data_dth_rcb.plot(legend=True,figsize=(18, 8), title='날짜별 전 세계 코로나 사망자,완치자 증가 변동 추이')
all2_cht.set_ylabel('Confirmed', fontsize=12)
```

```python
################### 2번#######################
# 나라별  확진자, 추이
country_data = data2.groupby(["Country/Region"]).sum()
country_data_sort = country_data.sort_values(by=['4/3/20'], axis=0,ascending=False)
```

```python
country_data_sort.drop(["Lat", "Long"], axis = 1, inplace = True)
country_data_sort = country_data_sort[:20]
country_data_sort
```

```python
country_data_total = country_data_sort[["4/3/20"]]
country_data_total
```

```python
#나라별 확진자 추이
plt.rc('font',family = 'Malgun Gothic')
con_cht = country_data_total.plot(kind='bar', legend=True,figsize=(18, 8), title='나라별 코로나 확진자 수 추이')
con_cht.set_ylabel('Confirmed', fontsize=12)
```

```python
######################3번#############################
#중국의 그래프
china_data = country_data_sort.loc[["China"]].T
china_data.columns = ['Confirmed']
china_data
```

```python
china_plot = china_data.plot(legend=True,figsize=(18, 8), title='날짜별 중국 코로나 확진자 수 추이')
china_plot.set_xlabel('Date', fontsize=12)
china_plot.set_ylabel('Confirmed', fontsize=12)
```

```python
#이탈리아아와 우리나라의 비교
twocon_data = country_data_sort.loc[["Italy","Korea, South"]].T
twocon_data.columns = ['Italy', 'Korea, South']
twocon_data
```

```python
twocon_plot = twocon_data.plot(legend=True,figsize=(18, 8), title='날짜별 중국 코로나 확진자 수 추이')
twocon_plot.set_xlabel('Date', fontsize=12)
twocon_plot.set_ylabel('Confirmed', fontsize=12)
```
