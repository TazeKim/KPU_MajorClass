```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.rc('font',family = 'Malgun Gothic')
```

```python
heights = [[1.6],[1.65],[1.7],[1.73],[1.8]]
```

```python
weights = [[60],[65],[72.3],[75],[80]]
```

```python
plt.title('키 대비 몸무게 그래프')
plt.xlabel('키(미터)')
plt.ylabel('몸무게(킬로그램)')
plt.plot(heights,weights,'k')
plt.axis([1.5,1.85,50,90])
plt.grid(True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X=heights,y=weights)
```

```python
weights = model.predict([[1.75]])[0][0]
```

```python
round(weights,2)
```

```python
heights = [[1.6],[1.65],[1.7],[1.73],[1.8]]
weights = [[60],[65],[72.3],[75],[80]]
plt.title('키 대비 몸무게 그래프')
plt.xlabel('키(미터)')
plt.ylabel('몸무게(킬로그램)')
plt.plot(heights,weights,'k')
plt.axis([1.5,1.85,50,90])
plt.grid(True)
plt.plot(heights,model.predict(heights),color ='r')
```

```python
plt.title('키 대비 몸무게 그래프')
plt.xlabel('키(미터)')
plt.ylabel('몸무게(킬로그램)')
plt.plot(heights,weights,'k')
plt.axis([0,1.85,-200,200])
plt.grid(True)
extreme_heights = [[0],[1.8]]
plt.plot(extreme_heights,model.predict(extreme_heights),color ='b')
```

```python
round(model.predict([[0]])[0][0],2)
```

```python
round(model.intercept_[0],2)
```

```python
round(model.coef_[0][0],2)
```

```python
import numpy as np
```

```python
print('RSS:%.2f'%np.sum((weights-model.predict(heights))**2))
```

```python
heights_test=[[1.58],[1.62],[1.69],[1.76],[1.82]]
```

```python
weights_test=[[58],[63],[72],[73],[85]]
```

```python
weights_test_mean=np.mean(np.ravel(weights_test))
```

```python
TSS=np.sum((np.ravel(weights_test)-weights_test_mean)**2)
```

```python
print("TSS:%.2f"%TSS)
```

```python
RSS = np.sum((np.ravel(weights_test)-np.ravel(model.predict(heights_test)))**2)
```

```python
print("RSS: %.2f"%RSS)
```

```python
R_squared=1-(RSS/TSS)
```

```python
print("R_squared:%.2f"%R_squared)
```

```python
print('R_squared:%.4f'%model.score(heights_test,weights_test))
```
