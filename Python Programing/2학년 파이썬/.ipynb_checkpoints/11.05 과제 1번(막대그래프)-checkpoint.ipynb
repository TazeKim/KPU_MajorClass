```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
```

```python
data = pd.read_csv('subwaystation.csv',encoding='euc-kr')
data = data[['호선명','승차승객수']]

df = data.groupby('호선명').sum()
df
```

```python
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)


ax = df.plot(kind='bar', title='호선별승객수', figsize=(10, 7), legend=True, fontsize=8,color=['red', 'green', 'blue', 'cyan', 'magenta'])
ax.set_xlabel('호선명', fontsize=8)          # x축 정보 표시
ax.set_ylabel('승차승객수', fontsize=8)     # y축 정보 표시
ax.legend(['승차승객수'], fontsize=8)    # 범례 지정
```
