```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = open('subwaytime.csv')
data = csv.reader(df)
next(data)
next(data)

df5 = []
df6 = []
df7 = []
sname = []
for row in data :
    row[4:] = map(int, row[4:])
    if row[1] == '4호선':
        df5.append(row[6])
        df6.append(row[8])
        df7.append(row[10])
        sname.append(row[3])


plt.rc('font',family = 'Malgun Gothic')
plt.title(' 4호선 승차인원수 변화율')
plt.plot(df5,label='5시')
plt.plot(df6,label='6시')
plt.plot(df7,label='7시')
plt.legend()
plt.xticks(range(26),sname,rotation = 90)
plt.show()
```
