```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
crime_data=pd.read_csv("crime_data.csv",encoding='cp949')
```

```python
crime_data.head()
```

```python
crime_station = crime_data.pivot_table(
    crime_data, index=["구분"], columns=["년도","죄종","발생검거"])

crime_station.head()
```

```python
crime_station = crime_station.xs('건수', axis=1, drop_level=True)
```

```python
crime_station.head()
```

```python
tmp=crime_station.columns.get_level_values(1) + crime_station.columns.get_level_values(2)
tmp
```

```python
date=list(crime_station.columns.get_level_values(0))
for i in range(len(date)):
    date[i]=str(date[i])# 문자열로 변환
```

```python
#컬럼을 이미 만들어준 두 컬럼으로 재설정
crime_station.columns=[date,tmp]
```

```python
crime_station.head()
```

```python
# 모든 해 검거 발생 총합
event=list(set(tmp))
event.sort() # 리스트 다시 정렬
date=list(set(date))
date.sort()
total=0
total_list = []
total_col = []
for i in event:
    for j in date:
        total+=crime_station[j][i].sum()
    
    print(i,': ',total)
    total_list.append(total)
    total_col.append(i)
    print('-----')
    total=0
```

```python
total_list
```

```python
total_col
```

```python
total_type = pd.DataFrame(total_list, index = total_col,columns = ['발생검거'])
total_type
```

```python
total_type_bal=total_type[1::2]#짝수번째 항만 모아서 만들기
type_ch = total_type_bal['발생검거']
```

```python
#################################### 쿼리문 2번##########################
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')

#categories = type_ch.index
#plt.pie(type_ch, labels=categories, autopct='%0.1f%%')
type_ch.plot(kind='pie', title='범죄 유형별 비중', textprops={'fontsize': 14})
plt.legend(loc='lower right')
plt.pie(type_ch,autopct='%1.1f%%')
plt.rcParams['figure.figsize'] = [10, 10]
```
