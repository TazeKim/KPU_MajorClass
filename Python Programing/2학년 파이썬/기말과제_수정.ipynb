```python
import numpy as np
import pandas as pd
from pandas import DataFrame as df
```

```python
crimedata = pd.read_csv('crime_data.csv',encoding='cp949')
crimedata.head()
```

```python
local=list(set(crimedata['구분']))
local
len(local)
range(len(local))
```

```python
#서울시 각 구의 총 범죄 발생 건수
a=0
j=0
for j in range(len(local)):
    print()
    a=0
    for i in range(2011,2019):
        seoul=crimedata[crimedata['구분']==local[j]]
        seoul=seoul[seoul['년도']== i]
        seoul=seoul[seoul['발생검거']=='발생']
        a += seoul['건수'].sum()
    print(local[j],':',a)
```

```python
#서울시 각 구의 총 범죄 검거 건수
y=0
j=0
print("서울시 각 구의 총 범죄 검거 건수")
for j in range(len(local)):
    print()
    y=0
    for i in range(2011,2019):
        seoul=crimedata[crimedata['구분']==local[j]]
        seoul=seoul[seoul['년도']== i]
        seoul=seoul[seoul['발생검거']=='검거']
        y += seoul['건수'].sum()
    print(local[j],':',y)
```

```python
#서울시 각 구의 총 범죄 검거율
a=0
j=0
y=0
for j in range(len(local)):
    print()
    a=0
    y=0
    for i in range(2011,2019):
        seoul=crimedata[crimedata['구분']==local[j]]
        seoul=seoul[seoul['년도']== i]
        seoul=seoul[seoul['발생검거']=='발생']
        a += seoul['건수'].sum()
        
        seoul=crimedata[crimedata['구분']==local[j]]
        seoul=seoul[seoul['년도']== i]
        seoul=seoul[seoul['발생검거']=='검거']
        y += seoul['건수'].sum()
        
    print(local[j],"검거율",':',round(y/a*100,2),'%')
```
