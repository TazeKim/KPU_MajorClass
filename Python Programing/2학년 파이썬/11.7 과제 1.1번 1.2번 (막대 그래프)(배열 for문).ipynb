```python
import csv
import matplotlib.pyplot as plt
```

```python
df = open('subwaytime.csv')
data = csv.reader(df)
next(data)
next(data)
max_in = [0] *24
max_name = [0]*24
for row in data :
    row[4:] = map(int, row[4:])
    for i in range(24) :
        a = row[i*2+4]
        if a >max_in[i]:
            max_in[i] = a
            max_name[i] = row[3] +'('+str(i+4)+'시)'

plt.rc('font',family = 'Malgun Gothic')
plt.bar(range(24),max_in)
plt.xticks(range(24),max_name,rotation = 90)
plt.show()

```

```python
df = open('subwaytime.csv')
data = csv.reader(df)
next(data)
next(data)
max_out = [0] *24
max_name = [0]*24
for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        b = row[j*2+5]
        if b >max_out[j]:
            max_out[j] = b
            max_name[j] = row[3] +'('+str(j+4)+'시)'

plt.bar(range(24),max_out)
plt.rc('font',family = 'Malgun Gothic')
plt.xticks(range(24),max_name,rotation = 90)
plt.show()
```
