```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('subwayfee.csv',encoding='euc-kr')
data=data[['호선명','유임승차','무임승차']]
df = data.groupby('호선명').sum()
udf= df[df['유임승차']==df['유임승차'].max()]
udf['유임승차']#유임승차 최대값
```

```python
mdf= df[df['무임승차']==df['무임승차'].max()]
mdf['무임승차']#무임승차 최대값
```
